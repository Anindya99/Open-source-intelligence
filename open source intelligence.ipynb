```python
df=pd.read_csv("../input/Militancy_News_Dataset.csv",index_col ="gname")
df.head()
```

```python
df_new = df.loc[['Jaish-e-Mohammad (JeM)','Lashkar-e-Taiba (LeT)','Hizbul Mujahideen (HM)']]
df_new.tail()
y=df_new.index
y.shape
df_new.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
v = TfidfVectorizer(stop_words='english')
x = v.fit_transform((df_new['addnotes']+df_new['scite1']+df_new['scite2']+df_new['scite3']).values.astype('U'))
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.75,random_state=42)
dummies=pd.get_dummies(y_train)
dummies
dummies2=pd.get_dummies(y_test)
dummies2.head()
```

## Using Logistic Regression

```python
for c in [0.01, 0.05, 0.25, 0.5, 1]:
    
    lr = LogisticRegression(C=c)
    lr.fit(x_train, y_train)
    print ("Accuracy for C=%s: %s" 
           % (c, accuracy_score(y_test, lr.predict(x_test))))
```

## When we are using each word as a single word and predicting through logistic regression highest accuracy is coming to be 55.84% for c=0.25 and c=1

## Support Vector Machine

```python
from sklearn.feature_extraction.text import CountVectorizer
ngram_vectorizer = CountVectorizer(binary=True, ngram_range=(1, 5),stop_words='english')
ngram_vectorizer.fit((df_new['addnotes']+df_new['scite1']+df_new['scite2']+df_new['scite3']).values.astype('U'))
x = ngram_vectorizer.transform((df_new['addnotes']+df_new['scite1']+df_new['scite2']+df_new['scite3']).values.astype('U'))
df_new = df.loc[['Jaish-e-Mohammad (JeM)','Lashkar-e-Taiba (LeT)','Hizbul Mujahideen (HM)']]
df_new.tail()
y=df_new.index
y.shape

x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.75,random_state=42)

for c in [0.01, 0.05, 0.25, 0.5, 1]:
    
    lr = LogisticRegression(C=c)
    lr.fit(x_train, y_train)
    print ("Accuracy for C=%s: %s" 
           % (c, accuracy_score(y_test, lr.predict(x_test))))
```
